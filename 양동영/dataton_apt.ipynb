day_care_type로는 등급을 측정하기 어려워 보여 숫자 데이터로만 등급 측정

```python
day_care_center.isnull().sum()
```

```python
day_care_center = day_care_center.fillna(0)
```

```python
# 교사 대 아동 비율
day_care_center['baby_to_teacher_score'] = day_care_center['day_care_baby_num'] / day_care_center['teacher_num']

# 아동 대 보육실 비율
day_care_center['baby_to_nursing_room_ratio'] = day_care_center['day_care_baby_num'] / day_care_center['nursing_room_num']

# 아동 대 놀이터 비율
day_care_center['baby_to_playground_ratio'] = day_care_center['day_care_baby_num'] / day_care_center['playground_num']

day_care_center
```

```python
# 어린이집 등급 측정 높을수록 좋음

# 교사 대 아동 비율 점수를 아동 대 교사 비율 점수로 변경 (더 낮은 비율은 더 좋은 것으로 가정)
day_care_center['baby_to_teacher_score'] = day_care_center['baby_to_teacher_score'].apply(lambda x: 0 if x == 0 else min(max((10 / x), 0), 10))

# 아동 대 보육실 비율 점수 계산 (더 낮은 비율은 더 좋은 것으로 가정)
day_care_center['baby_to_nursing_room_score'] = day_care_center['baby_to_nursing_room_ratio'].apply(lambda x: min(max((10 - x / 2), 0), 10))

# 아동 대 놀이터 비율 점수 계산 (더 낮은 비율은 더 좋은 것으로 가정)
day_care_center['baby_to_playground_score'] = day_care_center['baby_to_playground_ratio'].apply(lambda x: min(max((20 - x), 0), 10))

# 종합 점수 계산 (모든 비율 점수의 평균을 사용)
day_care_center['total_score'] = day_care_center[['baby_to_teacher_score', 'baby_to_nursing_room_score', 'baby_to_playground_score']].mean(axis=1)

# 등급 분류 (예: 총점에 따라 A, B, C 등급으로 분류)
day_care_center['grade'] = pd.cut(day_care_center['total_score'], bins=[0, 5, 7.5, 10], labels=['1', '2', '3'], right=True)

day_care_center
```
