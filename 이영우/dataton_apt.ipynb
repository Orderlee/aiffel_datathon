### 불필요한 컬럼 삭제

```python
df.drop(['학교ID','시도교육청코드','시도교육청명','교육지원청코드','운영상태','소재지도로명주소','교육지원청명','생성일자','변경일자','데이터기준일자','제공기관코드', '제공기관명'],axis=1, inplace=True)
```

### 컬럼명 영어로 변경

```python
df.rename(columns={
    '학교명': 'School_Name',
    '학교급구분': 'School_Level',
    '설립일자': 'Establishment_Date',
    '설립형태': 'Establishment_Type',
    '본교분교구분': 'Branch_Type',
    '소재지지번주소': 'Address',
    '위도': 'Latitude',
    '경도': 'Longitude'
}, inplace=True)
```

```python
df
```

```python
df = df[df['Address'].str.contains('부산광역시|서울특별시')]
df.reset_index(drop=True, inplace=True)
df
```

```python
df['Address'].str[:5]
```

### address 컬럼 분리

```python
# City, gu, dong으로 나누는 함수
def split_address(address):
    
    parts = address.split(' ')
    city = parts[0]
    gu = parts[1]
    dong = ''.join(parts[2])
    return city, gu, dong

df[['city', 'gu', 'dong']] = df['Address'].apply(lambda x: pd.Series(split_address(x)))

print(df[['city', 'gu', 'dong']])
```

```python
df_columns = ['School_Name', 'School_Level', 'Establishment_Date', 'Establishment_Type', 'Branch_Type', 'Address', 'city', 'gu', 'dong', 'Latitude', 'Longitude', ]
df = df[df_columns]
df
```

### CSV 저장

```python
df.to_csv(home+'school_data.csv')
```
